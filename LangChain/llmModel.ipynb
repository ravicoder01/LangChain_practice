# Chat Prompt (instruct model to behave like something)

```python
from langchain_core.prompts import ChatPromptTemplate

prompt= ChatPromptTemplate.from_messages(
    [
         ("system", "You are an AI expert engineer. Provide me answer based on question"),
         ("user", "{input}")
    
    ]
)
prompt
```

```python
chain= prompt| llm
result= chain.invoke({"input": "what is the use of capacitor?"})
result
```

**Using output parser to generate only needed output**

```python
from langchain_core.output_parsers import StrOutputParser
response= StrOutputParser()
chain= prompt| llm| response
chain.invoke("Explain the core difference between ML and DL")
```
